```python
from llama_cpp import Llama

model_file = "Qwen2.5-32B-Instruct-Q2_K.gguf"

print("Initiating direct VRAM load. Bypassing System RAM...")
llm = Llama(
    model_path=f"/content/models/{model_file}",
    n_gpu_layers=-1,
    n_ctx=1024,      # Strictly locked at 1024 to prevent KV cache overflow
    use_mmap=False,  # The critical bypass flag
    verbose=True
)

# Dummy prompt to stress Prompt Processing
prompt = "Analyze the memory bandwidth constraints of a T4 GPU. " * 60

print("\n--- Starting 32B Stress Test ---")
output = llm(
    prompt,
    max_tokens=128,
)

print("\nStress test complete! Check the stderr output.")
```
